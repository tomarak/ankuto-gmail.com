```python
mushroom_types = 'amanita_muscaria','amanita_phalloides','galerina_marginata', 'gyromitra_esculenta', 'pleurocybella_porrigens', 'conocybe_filaris'
path = Path('mushrooms')
```

```python
for o in mushroom_types:
    dest = (path/o)
    if dest.exists():
        continue
    dest.mkdir(exist_ok=True)
    results = search_images_bing(key, f'{o.replace("_", " ")}')
    print(results)
    download_images(dest, urls=results.attrgot('content_url'))
```

```python
fns = get_image_files(path)
fns
```

```python
failed = verify_images(fns)
failed
```

```python
failed.map(Path.unlink);
```

## From data to DataLoaders

```python
mushrooms = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.3, seed=42),
    get_y=parent_label,
    item_tfms=Resize(128))
```

```python
dls = mushrooms.dataloaders(path)
```

```python
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
mushrooms = mushrooms.new(item_tfms=Resize(128, ResizeMethod.Squish))
dls = mushrooms.dataloaders(path)
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
mushrooms = mushrooms.new(item_tfms=Resize(128, ResizeMethod.Pad, pad_mode='zeros'))
dls = mushrooms.dataloaders(path)
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
mushrooms = mushrooms.new(item_tfms=RandomResizedCrop(128, min_scale=0.3))
dls = mushrooms.dataloaders(path)
dls.train.show_batch(max_n=4, nrows=1)
```

### Data augmentation

```python
mushrooms = mushrooms.new(item_tfms=Resize(128), batch_tfms=aug_transforms(mult=2))
dls = mushrooms.dataloaders(path)
dls.train.show_batch(max_n=8, nrows=2)
```

## Training the model

```python
mushrooms = mushrooms.new(
    item_tfms=RandomResizedCrop(224, min_scale=0.5),
    batch_tfms=aug_transforms())
dls = mushrooms.dataloaders(path)
```

```python
learn = cnn_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(5, nrows=5)
```

```python
cleaner = ImageClassifierCleaner(learn)
cleaner
```

```python
for idx in cleaner.delete(): cleaner.fns[idx].unlink()
for idx,cat in cleaner.change(): shutil.move(str(cleaner.fns[idx]), path/cat)
```

### Creating a Notebook app from the model

```python
btn_upload = widgets.FileUpload()
btn_upload
```

```python
img = PILImage.create(btn_upload.data[-1])
```

```python
out_pl = widgets.Output()
out_pl.clear_output()
with out_pl: display(img.to_thumb(128,128))
out_pl
```

```python
pred,pred_idx,probs = learn_inf.predict(img)
```

```python
lbl_pred = widgets.Label()
lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
lbl_pred
```

```python
btn_run = widgets.Button(description='Classify')
btn_run
```

```python
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)
```

```python
btn_upload = widgets.FileUpload()
```
